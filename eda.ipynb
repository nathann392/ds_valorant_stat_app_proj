### <b>Agents</b>

```python
df1 = df[['top_1_agents_name', 'top_1_agents_hours', 'top_1_agents_win%', 'top_1_agents_acs']]
df1 = df1.rename(columns={'top_1_agents_name': 'name', 'top_1_agents_hours': 'hours', 'top_1_agents_win%': 'win%', 'top_1_agents_acs': 'acs'})

df2 = df[['top_2_agents_name', 'top_2_agents_hours', 'top_2_agents_win%', 'top_2_agents_acs']]
df2 = df2.rename(columns={'top_2_agents_name': 'name', 'top_2_agents_hours': 'hours', 'top_2_agents_win%': 'win%', 'top_2_agents_acs': 'acs'})

df3 = df[['top_3_agents_name', 'top_3_agents_hours', 'top_3_agents_win%', 'top_3_agents_acs']]
df3 = df3.rename(columns={'top_3_agents_name': 'name', 'top_3_agents_hours': 'hours', 'top_3_agents_win%': 'win%', 'top_3_agents_acs': 'acs'})
```

```python
agent_df = pd.DataFrame(columns=['name', 'hours', 'win%', 'acs'])
agent_df = pd.concat([df1, df2, df3], ignore_index=True)
```

```python
agent_df.groupby('name')[['win%', 'acs']].mean()
```

```python
agent_df.groupby('name')[['hours']].sum()
```

```python
plt.figure(figsize=(16,4))
ax = sns.barplot(data = agent_df, x='name', y='hours', estimator=sum, errwidth=0)
a = ax.set(title='Hours vs. Agent')
b = ax.bar_label(ax.containers[0])
```

```python
plt.figure(figsize=(16,4))
ax = sns.barplot(data = agent_df, x='name', y='acs', errwidth=0)
a = ax.set(title='Score vs. Agent')
b = ax.bar_label(ax.containers[0])
```

```python
plt.figure(figsize=(16,4))
ax = sns.barplot(data = agent_df, x='name', y='win%', errwidth=0)
a = ax.set(title='Win % vs. Agent')
b = ax.bar_label(ax.containers[0])
```

```python
controller = ['Astra', 'Brimstone', 'Viper', 'Harbor', 'Omen']
initiator = ['KAY/O', 'Sova', 'Fade', 'Skye', 'Breach']
sentinel = ['Chamber', 'Sage', 'Cypher', 'Killjoy']
duelist = ['Jett', 'Reyna', 'Raze', 'Yoru', 'Phoenix', 'Neon']

def role_simplifier(role):
    if role in controller:
        return 'Controller'
    elif role in initiator:
        return 'Initiator'
    elif role in sentinel:
        return 'Sentinel'
    elif role in duelist:
        return 'Duelist'
    else:
        return 'N/A'
```

```python
agent_df['role'] = agent_df['name'].apply(role_simplifier)
```

```python
# Role vs. Average of Score and Win %
pd.pivot_table(agent_df, index='role', values=['acs','win%'])
```

```python
# Role vs. Sum of Hours
pd.pivot_table(agent_df, index='role', values=['hours'], aggfunc='sum')
```

```python
controller_df = agent_df[agent_df.name.isin(controller)]
initiator_df = agent_df[agent_df.name.isin(initiator)]
sentinel_df = agent_df[agent_df.name.isin(sentinel)]
duelist_df = agent_df[agent_df.name.isin(duelist)]

for i in [controller_df, initiator_df, sentinel_df, duelist_df]:
    plt.figure(figsize=(6,3))
    ax = sns.barplot(data=i, x='name', y='hours', estimator=sum, errwidth=0)
    a = ax.bar_label(ax.containers[0])
    plt.show()
```

```python
for i in [controller_df, initiator_df, sentinel_df, duelist_df]:
    plt.figure(figsize=(6,3))
    ax = sns.barplot(data=i, x='name', y='win%', errwidth=0)
    a = ax.bar_label(ax.containers[0])
    plt.show()
```
